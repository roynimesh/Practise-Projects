```python
def ordinal_encode(df,column,ordering):
    df=df.copy()
    df[column]=df[column].apply(lambda x:ordering.index(x))
    return df
```

```python
def onehot_encode(df,column):
    df=df.copy()
    dummies=pd.get_dummies(df[column])
    df=pd.concat([df,dummies],axis=1)
    df.drop(column, axis=1, inplace=True)
    return df
```

```python
df=ordinal_encode(df,'Date',date_ordering)
```

```python
from sklearn.model_selection import cross_val_score
def regression(model,x,y):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=10)
    model.fit(x_train,y_train)
    print('Accuracy :', model.score(x_test,y_test)*100)
    score=cross_val_score(model,x,y,cv=8)
    print("CV Score", np.mean(score)*100)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
regression(model,x,y)
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
regression(model,x,y)
```

# For Classification

```python
dflr=pd.read_csv('avocado.csv')
dflr
```

```python
date_order=sorted(dflr['Date'].unique())
```

```python
def ordinal_encode(dflr,column,order):
    dflr=dflr.copy()
    dflr[column]=dflr[column].apply(lambda x:order.index(x))
    return dflr
```

```python
dflr=ordinal_encode(dflr,'Date',date_order)
```

```python
dflr
```

```python
type=pd.get_dummies(dflr['type'],drop_first=True)
type
```

```python
dflr.drop(['type'],axis=1,inplace=True)
dflr.head()
```

```python
dflr=pd.concat([dflr,type],axis=1)
dflr.head()
```

```python
dflr.drop('Unnamed: 0',axis=1)
```

```python
features=dflr.drop('region',axis=1)
target=dflr['region']
```

```python
scaler=StandardScaler()
features=scaler.fit_transform(features)
```

```python
features_train,features_test,target_train,target_test=train_test_split(features,target,test_size=0.20,random_state=42)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.ensemble import RandomForestClassifier
f=RandomForestClassifier(n_estimators=20, criterion='entropy',random_state=42)
f.fit(features_train,target_train)
```

```python
from sklearn.metrics import confusion_matrix
m=confusion_matrix(target_test,f.predict(features_test))
TN=m[0][0]
TP=m[1][1]
FN=m[1][0]
FP=m[0][1]
print(m)
print('Accuracy = {}'.format((TP+TN)/(TP+TN+FP+FN)))
```

```python
score=cross_val_score(f,features,target,cv=5)
print("CV Score of RandmForestClassifier :", np.mean(score)*100)
```

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression(max_iter=10000)
lg.fit(features_train,target_train)
```

```python
cm=confusion_matrix(target_test,lg.predict(features_test))
TN=cm[0][0]
TP=cm[1][1]
FN=cm[1][0]
FP=cm[0][1]
print(cm)
print('Accuracy = {}'.format((TP+TN)/(TP+TN+FP+FN)))
```

```python
score=cross_val_score(lg,features,target,cv=5)
print("CV Score of LogisticRegression :", np.mean(score)*100)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier(criterion='entropy')
dtc.fit(features_train,target_train)
```

```python
dcm=confusion_matrix(target_test,dtc.predict(features_test))
TN=dcm[0][0]
TP=dcm[1][1]
FN=dcm[1][0]
FP=dcm[0][1]
print(dcm)
print('Accuracy = {}'.format((TP+TN)/(TP+TN+FP+FN)))
```

```python
score=cross_val_score(dtc,features,target,cv=5)
print("CV Score of DecisionTreeClassifier :", np.mean(score)*100)
```
