```python
df=df.drop('Over18',axis=1)
df=df.drop('EmployeeNumber',axis=1)
df=df.drop('StandardHours',axis=1)
df=df.drop('EmployeeCount',axis=1)
df
```

```python
plt.figure(figsize=(16,14))
sns.heatmap(df.corr(),annot=True,annot_kws={"size":12},cmap='YlGnBu',vmin=-1.0,vmax=1.0)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
for column in df.columns:
    if df[column].dtype==np.number:
        continue
    df[column]=LabelEncoder().fit_transform(df[column])
```

```python
df
```

```python
x=df.drop(columns=['Attrition'])
y=df['Attrition']
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
f=RandomForestClassifier(n_estimators=20, criterion='entropy',random_state=42)
f.fit(x_train,y_train)
```

```python
f.score(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix
m=confusion_matrix(y_test,f.predict(x_test))
TN=m[0][0]
TP=m[1][1]
FN=m[1][0]
FP=m[0][1]
print(m)
print('Accuracy = {}'.format((TP+TN)/(TP+TN+FP+FN)))
```

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression(max_iter=10000)
lg.fit(x_train,y_train)
```

```python
f.score(x_train,y_train)
```

```python
cm=confusion_matrix(y_test,lg.predict(x_test))
TN=cm[0][0]
TP=cm[1][1]
FN=cm[1][0]
FP=cm[0][1]
print(cm)
print('Accuracy = {}'.format((TP+TN)/(TP+TN+FP+FN)))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier(criterion='entropy')
dtc.fit(x_train,y_train)
```

```python
f.score(x_train,y_train)
```

```python
dcm=confusion_matrix(y_test,dtc.predict(x_test))
TN=dcm[0][0]
TP=dcm[1][1]
FN=dcm[1][0]
FP=dcm[0][1]
print(dcm)
print('Accuracy = {}'.format((TP+TN)/(TP+TN+FP+FN)))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score=cross_val_score(f,x,y,cv=5)
print("CV Score of RandmForestClassifier :", np.mean(score)*100)
```

```python
score=cross_val_score(lg,x,y,cv=5)
print("CV Score of LogisticRegression :", np.mean(score)*100)
```

```python
score=cross_val_score(dtc,x,y,cv=5)
print("CV Score of DecisionTreeClassifier :", np.mean(score)*100)
```

```python
y_pred=lg.predict(x_test)
```
